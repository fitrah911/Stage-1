## Univariate Analysis

Boxplot

```python
for i in range(0, len(nums)):
    plt.subplot(2, 8, i+1)
    sns.boxplot(y=df_sumber[nums[i]], color='gray', orient='v')
    plt.tight_layout()
```

Semua kolom kecuali kolom `agent` dan `company` memiliki outlier maka akan dilakukan proses penghapusan outlier

Distribution Plot

```python
for i in range(0, len(nums)):
    plt.subplot(2, 8, i+1)   # (baris, kolom, posisi)
    sns.distplot(df_sumber[nums[i]], color='gray')
    plt.tight_layout()
```

Semua kolom bernilai negatif skewed atau skewed kiri maka perlu dilakukan normalisasi agar distribusinya menjadi normal

## Multivariate Analysis

```python
df_sumber.corr()
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df_sumber.corr(), cmap='Blues', annot=True, fmt='.2f')
```

Dari Heatmap diatas, dapat disimpulkan :
1. Semua data relevan, sehingga tidak ada data yang di drop kecuali kolom yang membuat multicollinearity
2. kolom `stays_in_week_nights` dan `stays_in_weekend-nights` yang memiliki nilai 0,5 menyebabkan multicollinearity, sehingga salah satu dari kolom akan di drop
3. Kolom `arrival_date_year` dan `arrival_date_week_number` yang memiliki nilai -0,54 menyebabkan multicollinearity, sehingga salah satu dari kolom akan di drop
